### Лист 3, задача 13


###### Условие: 

У тети Маши — двое детей, один старше другого. Предположим, что вероятности рождения
мальчика и девочки равны и не зависят от дня недели, а пол первого и второго ребенка
независимы. Для каждой из четырех ситуаций найдите условную вероятность того, что у
тёти Маши есть дети обоих полов.

(a) Известно, что хотя бы один ребенок — мальчик.

(b) Тетя Маша наугад выбирает одного своего ребенка и посылает к тете Оле, вернуть
учебник по теории вероятностей. Это оказывается мальчик.

(c) Известно, что старший ребенок — мальчик.

(d) На вопрос: «А правда ли тетя Маша, что у вас есть сын, родившийся в пятницу?»
тётя Маша ответила: «Да».



###### Решение:

Введем необходимые обозначения для удобства

B - мальчик

G - девочка

R - разнополые дети

(a) 

Тут все совсем просто. Может быть 4 исхода: $BG$, $BB$, $GG$, $GB$. Объединение событий $R$ и $B\ge1$ - это рождение разнополых детей. Вероятность данного события, очевидно, равна $\frac{1}{2}$. 3 из 4 исходов удовлетворяют условию $B\ge1$. Отсюда:

$P(R|B\ge1)=\frac{P(R \cup (B\ge1))}{P(B\ge1)}=\frac{\frac{1}{2}}{\frac{1}{4}*3}=\frac{2}{3}$

(b)

Пусть:

rB - случайный (random) boy
![](http://s013.radikal.ru/i322/1610/38/05bc3cdaffb4.png)
$P(R|rB)=\frac{P(R \cup rB)}{P(rB)}=\frac{\frac{1}{8}+\frac{1}{8}}{\frac{1}{8}+\frac{1}{8}+\frac{1}{8}}=\frac{1}{2}$

(c)

Пусть:

FB - старший (first) boy

$P(R|FB)=\frac{P(R \cup FB)}{P(FB)}=\frac{\frac{1}{2}}{\frac{1}{4}*3}=\frac{2}{3}$

(d)

Пусть:

PB - Пятничный (pyatnichniy) boy

NP - родился не в пятницу (nye pyathichniy)
![](http://s019.radikal.ru/i641/1610/dc/02899223e297.png)
$P(R|PB)=\frac{P(R \cup PB)}{P(PB)}=\frac{\frac{1}{28}+\frac{1}{28}}{\frac{1}{14}+\frac{3}{98}+\frac{1}{28}}=\frac{14}{27}$




## Функция плотности

Позволю себе написать небольшую вводную часть:

Функция $f(x)$ называется функцией плотности распределения вероятностей непрерывной случайной величины, если она является производной от функции распределения.

Следовательно, вероятность попадания случайной величины в диапазон $[a,b]$ вычисляется как площадь под графиком функции плотности на данном промежутке![](https://upload.wikimedia.org/wikipedia/commons/thumb/f/f2/Integral_as_region_under_curve.svg/280px-Integral_as_region_under_curve.svg.png)


### Придуманная задача

Пусть X - случайная величина
Функция плотности f(t) задана таким рисунком: ![](http://s015.radikal.ru/i330/1610/91/b8f0b3036e89.png)

а) Записать формулой

б) 

$P(x>2)$
$P(x<2)$
$P(x=3)$

###### Решение

a) $f(t)=\left\{ 
\begin{eqnarray}
0, t\in (-\infty;0)\cup(4;+\infty)\\
\frac{1}{8}t, t\in[0;4].\\
\end{eqnarray}
\right.$

б) 

$P(X>2)=\int\limits_2^4\frac{1}{8}tdt=\frac{3}{4}$

Верояность можно вычислять не обязательно через интеграл, но и через площадь:
$P(X<2)=\frac{2*\frac{1}{4}}{2}=\frac{1}{4}$

$P(X=3)=\int\limits_3^3\frac{1}{8}tdt=0$

Обывателя может обескуражить тот факт, что вероятность возможного события ($x=3$) равна нулю. 

Но нам важно понимать, что если вероятность нулевая, это не значит, что событие невозможо. (А вот если оно невозможно, то вероятность нулевая точно, тут все работает)


### Лист 6, задача 1
###### Условие
Функция плотности случайной величины $X$ равна $5$ при $x = 7$. Найдите примерно вероятность того, что $X$ попадёт в отрезок $[7; 7.001]$.
###### Решение
$P(x\in[7;7.001])\approx5*0.001=0.005$

Естественно, мы не знаем, как точно ведет себя функция плотности, поэтому ответ примерный.

### Лист 6, задача 2
###### Условие 
Случайная величина $X$ имеет функцию плотности $f$. С помощью $o(\Delta)$ и $f(x)$ запишите
вероятность $P(X ∈ [x; x + ∆])$.

###### Решение

$o(\Delta)$ стремится к $0$ больше, чем $\Delta$

$\Delta*f(x)$ это часть площади под графиком функции плотности  на промежутке $[x;x+\Delta]$, но не вся, к ней следует прибавить величину, меньшую $\Delta$
![](http://i069.radikal.ru/1610/38/d404a5b209d4.png)
Следовательно, имеем:

$P(X\in[x;x+\Delta]=\Delta*f(x)+o(\Delta)$
