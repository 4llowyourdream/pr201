## Упражнения по теме.

Случайная величина $X$ имеет нормальное распределение с математическим ожиданием $\mu$ и дисперсией $\sigma^2$:

$X\sim N(\mu,\sigma^2)$

### Формула нормального распределния: 

$\Large f(x)=\frac{1}{\sqrt{2\pi\sigma^2}}e^{-\frac{(x-\mu)^2}{2\sigma^2}}$

### Упражнение 2:

$Y_1,Y_2,Y_3$ - независимы

$Y_i\sim N[0,6]$,  $\forall i$

1. Найти $E(Y_i)$, $Var(Y_i)$

2. Если $S=Y_1+Y_2+...+Y_{100}$, то найти $P(S>330)$

3. Найти $t$ такое, что $P(S<t)=0.7$ (на дом)

Решение:
1. $\large E(Y_i)=3$

   $\large Var(Y_i)=E(Y_i^2)-E(Y_i)^2=12-9=3$, т.к. 
   
   $\large E(Y_i^2)=\int_{0}^{6}Y_i^2f(Y_i)dy=\frac{Y_i^3}{18}=12$

2. $\large E(S)=E(Y_1)+E(Y_2)+...+E(Y_{100})=100E(Y_i)=300$

   $\large Var(S)=\sum_{i=1}^{100}Var(Y_i)=100Var(Y_i)=300$
   
   Теперь сделаем равносильное преобразование, с учетом того, что $S\sim N(300,300)$:
   
   $\large S'=\frac{S-E(S)}{\sqrt{Var(S)}}$
   
   Тогда имеем:
   
   $\large Var(S')=Var\bigg(\frac{S-E(S)}{\sqrt{Var(S)}}\bigg)=\frac{Var(S)}{Var(S)}=1$
   
   Значит, $S'$ - нормальное распределение с параметрами $\mu=0$, $\sigma^2=1$.
   
   Тогда:
   
   $\large P(S>330)=P\bigg(\frac{S-300}{\sqrt{300}}>\frac{330-300}{\sqrt{300}}\bigg)=P(S'>1.73)$
