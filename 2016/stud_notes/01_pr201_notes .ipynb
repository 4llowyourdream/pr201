## Задача 1

#### Условие:

В вазе 5 неотличимых с виду конфет: Две без ореха и три - с орехом. Маша ест конфеты, выбирая их наугад до тех пор, пока не съест первую конфету с орехом. 

#### Решение:

Пусть X - количество съеденных Машей конфет: Y - количество съеденных конфет с орехом. 
$P(X=1)=?$ Найдем вероятность того, что Маша остановится на 1 конфете. Для того, чтобы найти вероятности, надо нарисовать дерево возможных событий : так будет нагляднее и понятнее. 
![](http://s017.radikal.ru/i414/1610/59/6d5115e72875.png)

Из дерева видно, что Маша закончит на первом ходе в том случае, если первая конфета, которую она выберет, окажется с орехом. Это событие случается с вероятностью $\dfrac{3}{5}$ (всего конфет 5, а с орехом из них 3, Маша выбирает конфету один раз). Значит $P(X=1)=\dfrac{3}{5}$ 


Маша остановится на второй конфете, если вторая конфета, которую она выберет, будет с орехом. Рассмотрим то, как это получится. Певая конфета должна быть без ореха, а вероятность этого равна $\dfrac{2}{5}$ . Вторая конфета должна быть с орехом. Всего осталось 4 конфеты, из них с орехом - 3. Значит $P(II-c-орехом)=\dfrac{3}{4}$. Чтобы найти $P(X=2)$, надо умножить вероятности выпадения такой последовательности, так как события независимые: $P(X=2)=\dfrac{2}{5} \cdot \dfrac{3}{4}=\dfrac{3}{10}$

Найдем вероятность того, что Маша остановится на третьей конфете. По картинке видно, что к третьему ходу останется 3 конфеты, т.к. 2 конфеты уже были извлечены. И раз на них Маша не остановилась, то они были без ореха. Всего конфет с орехом осталось 3 штуки, и в корзине лежит 3 конфеты, следовательно, вероятность того,что Маша на третий раз извлечет конфету с орехом равна 1. Значит,  $P(X=3)=\dfrac{2}{5} \cdot \dfrac{1}{4}\cdot 1 = \dfrac{1}{10}$.

Теперь, когда мы знаем вероятности $P(X=2)$ и $P(X=3)$, мы можем найти $P(X>1)=P(X=2)+ P(X=3)= \dfrac{3}{10}+ \dfrac{1}{10}= \dfrac{4}{10}$. Складываем мы их потому что надо найти X>1, а в нашей задаче это будет либо X=2, либо X=3. Если мы говорим "либо", и если события не могут произойти совместно(к примеру, за раз получиться и Х=2, и Х=3), то вероятности надо сложить по следующей теореме. 

Теорема. Вероятность суммы конечного числа несовместных событий равна сумме вероятностей этих событий. Более полная формулировка теоремы и доказательство доступны по ссылке : http://www.grandars.ru/student/vysshaya-matematika/slozhenie-i-umnozhenie-veroyatnostey.html

$E(X)$-(Expected value) математическое ожидание. Среднее значение переменной при стремлении количества случаев к бесконечности. Мы находим его, сложив произведения значений $x_i$, которые принимает случайная величина $X$, на соответствующие вероятности $Р$. 

$E(X)= 0,6 \cdot 1 + 0,3 \cdot 2 + 0,1 \cdot 3= 1,5$ .



## Задача 6

#### Условие:

Саша и Маша по очереди подбрасывают кубик до первой шестерки. Посуду будет мыть тот, кто первым выбросит шестерку. Маша бросает кубик первой. Какова вероятность того, что посуду будет мыть Маша? Сколько в среднем раз они будут бросать кубик?

#### Решение:

Пусть $X$ - количество бросков кубика. 

Построим дерево.
![](http://s19.radikal.ru/i192/1610/66/8b3d882b9659.png)

Так как начинает бросать Маша, то, если в первом же шаге(нечётный) выпадет $6$, она пойдет мыть посуду. Затем бросает Саша, и если у него выпадет $6$, то он будет мыть. Потом снова бросает Маша и так далее. Следовательно, если $X$ - нечётное, то посуду будет мыть Маша, а если чётное -  то Саша. Фразу "вероятность того, что посуду будет мыть Маша" можно сформулировать на математическом языке как $P(X-чётное)$.

Сначала найдем вероятность того, что игра закончится на первом ходе, т.е. сразу выпадет $6$. Всего исходов 6, а выпадению шестерки соответствует один из них, значит $P(X=1)=\dfrac{1}{6}$.

Чтобы найти $P(X=2)$, перемножим вероятности событий, которые должны случиться последовательно, а именно: первым выпадает "не 6", во втором броске выпадает "6".$P(X=2)=\dfrac{5}{6} \cdot \dfrac{1}{6}=\dfrac{5}{36}$

Дерево наглядно показывает следующее: если у Маши выпала не шестерка, и у Саши выпала не шестерка, то получается ничья, и фактически игра и не начиналась. Следовательно, $P(X-чётное)=\dfrac{1}{6}+\dfrac{5}{6}\cdot \dfrac{5}{6}\cdot P(X-чётное)$. Первое слагаемое означает то, что Маша пойдет мыть после первого броска. Второе слагаемое возвращает нас к первому ходу в том случае, если Саша не выбросил шестерку( и игра вернулась в начальное положение, за  исключением того, что уже был сделан один ход, который мы учли как первое слагаемое). Решаем уравнение и находим $P(X-чётное)=\dfrac{6}{11}$

Теперь найдем математическое ожидание, сколько в среднем раз они будут бросать кубик.  
$E(X)=P(X=1)\cdot 1+P(X=2) \cdot 2 + P(X=3) \cdot 3 + P(X=4) \cdot 4 ...=$ <tex>$\sum\limits_{i=1}^n P(X=i)\cdot i$</tex> = $\dfrac{1}{6} \cdot 1 + \dfrac {5}{6}\cdot \dfrac{1}{6} \cdot 2+\dfrac {5}{6} \cdot \dfrac {5}{6}\cdot \dfrac{1}{6} \cdot 3 + \dfrac {5}{6} \cdot \dfrac {5}{6} \cdot \dfrac {5}{6}\cdot \dfrac{1}{6} \cdot 4 + ...  $

Запишем, используя степени : $E(X)=\dfrac{1}{6} \cdot 1 + \dfrac {5}{6^2}\cdot 2+\dfrac {5^2}{6^3} \cdot 3 + \dfrac {5^3}{6^4} \cdot 4 + ... =$ <tex>$\sum\limits_{k=1}^n \frac{5^{k-1}}{6^k}\cdot k$</tex> = $S$ 

Теперь домножим эту сумму , которую мы обозначили за $S$, на $\dfrac{5}{6}$ , чтобы получить бесконечно убывающую геометрическую прогрессию : $\dfrac{5}{6}\cdot S= \dfrac{5}{6} \cdot \dfrac{1}{6} \cdot 1 + \dfrac {5^2}{6^3}\cdot 2+\dfrac {5^3}{6^4} \cdot 3 + \dfrac {5^4}{6^5} \cdot 4 + ...  $

$ S-\dfrac{5}{6} \cdot S= \dfrac{1}{6} \cdot 1 + \dfrac {5}{6^2}\cdot 1+\dfrac {5^2}{6^3} \cdot 1 + \dfrac {5^3}{6^4} \cdot 1 + ... $ . В полученной бесконечно убывающей геометрической прогрессии,  $q=\dfrac{5}{6}$, $b_1=\dfrac{1}{6}$.

$ S-\dfrac{5}{6} \cdot S= \dfrac{1}{6} \cdot S= \dfrac {b_1}{1-q}= \dfrac{\frac{1}{6}}{1-\frac{5}{6}}=1 $ Следовательно, $E(X)=S=6$

Ответ :  $P$(Маша моет)$=\dfrac{6}{11} , E(X)=6$
