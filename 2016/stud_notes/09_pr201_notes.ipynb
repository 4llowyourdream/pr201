## Цыганка vs  Loch (Ness Monster)
![](https://pp.vk.me/c626828/v626828933/3641f/003Nr8fFDdE.jpg)

Всего в колоде 52 карты. Студенты выбирают у гадалки Насти 5 карт, затем она прячет одну карту, остальные расставляет в особом порядке. Угадает ли Loch карту?

Первый расклад карт: 6&#9830;, 7&#9824;, Q&#9829;, 4&#9827;. Загаданная карта: 6&#9827;

Второй расклад карт: 2&#9829;,J&#9829;,A&#9829;,7&#9829;. Загаданная карта: 8&#9829;

Третий расклад карт: Q&#9830;,Q&#9829;,Q&#9827;,Q&#9824;. Загаданная карта:4&#9830;

Каждый раз Loch отгадывает карту.

Решение:

Пять карт разложены цыганкой в особом порядке. 

Следовательно, среди них точно есть одномастная пара, так как карт 5, а мастей 4. Если такая пара всего одна, то карта загадывается среди них. Как отгадать карту? 

Всего карт одной масти — 13. Расставим их по кругу. Максимальное расстояние между двумя картами— 6 карт, которые могут быть расставлены 3! =6 способами. 

Далее в силу вступает договорённость между цыганкой и Loch-нэсом. Одна из расставленных четырёх карт (в данном случае самая правая) отвечает за масть загаданной карты. Остальные три расставляются согласно ранее оговорённым манёврам, которые обозначают расстояние между парой одномастных карт. (Отгадке обязаны Александру Тащяну и Ярославу Марченко)

## Немного теории

![](https://pp.vk.me/c636728/v636728256/494ac/PQ_1Jj997ns.jpg)


На этом множестве существует геометрия: у случайных величин можно померить угол между ними, длину. 

Опеределение 1: длина случайной величины X равняется $\sigma_{x}=\sqrt{Var(x)}$

Опеределение 2: $\cos(X:Y)=corr(X;Y)$

Упражнение: Монетка подбрасывается два раза. X –  количество орлов, 

$\begin{equation*}
Y = 
 \begin{cases}
   1 \text{, если орёл был в первом броске}\\
   0 \text{, иначе}
 \end{cases}
\end{equation*}
$

$
\begin{equation*}
Z = 
 \begin{cases}
   1 \text{, если орёл был во втором броске}\\
   0 \text{, иначе}
 \end{cases}
\end{equation*}$

Найти длину X, Y, Z

а) $corr(X,Y)=\cos(X,Y)$, т.е. частичную корреляцию X и Y при фиксированном Z$

б) $\cos(\tilde{X},\tilde{Y})$

Решение:

![](https://pp.vk.me/c636728/v636728256/494b4/goRWl2tQq9o.jpg) 


а) Запишем исходы выпадения монет при подбрасывании два раза:

$P(\text{Орёл-Орёл})=1/4$

$P(\text{Орёл-Решка})=1/4$

$P(\text{Решка-Орёл})=1/4$

$P(\text{Решка-Решка})=1/4$

$X={0;1;2} E(X)=1/2+2*1/4=1$

$X^{2}={0;1;4} $ $ E(X^{2})=1/2+4*1/4=3/2$

$D(X)=E(X^{2})-E(X)^{2}=1/2$

$E(Y)=1/2$ $ E(Y^{2})=1/2 D(Y)=1/2-1/4=1/4$

$E(Z)=1/2$ $ E(Z^{2})=1/2 D(Z)=1/2-1/4=1/4$

$\sigma_{X}=\sqrt{X}=\sqrt{2}/2$

$\sigma_{Z}=\sigma_{Y}=1/2$

$\alpha=\dfrac{Cov(X,Y)}{\sigma_{X}\sigma_{Y}}$, где $\alpha$ – угол между X и Y

$Cov(X,Y)=E(XY)-E(X)E(Y)=2*1/4+1/4-1/2=1/4$

$Corr(X,Y)=\dfrac{1/4}{1/\sqrt{2}*1/2}=\dfrac{2\sqrt{2}}{2} \Rightarrow \alpha=\pi/4$

b)$\tilde{X}=X-\alpha Z$

По теореме Пифагора:

$Var(\tilde{X})+Var(\alpha Z)=Var(X)$

$X-\alpha Z\perp Z \Rightarrow Cov(X-\alpha Z,Z)=0$

$Cov(X,Z)-Cov(\alpha Z,Z)=0 $

$cos(\alpha)=\dfrac{Cov(X,Z)}{Var(Z)}=1$

$\tilde{Y}=Y$ , Z и Y независимы, $ Z\perp Y \Rightarrow  Cov(Z,Y)=0. \Rightarrow \cos(\tilde{X},\tilde{Y})=1$
