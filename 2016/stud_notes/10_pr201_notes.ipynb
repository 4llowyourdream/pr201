## Упражнение

Пусть даны две случайные величины $X$ и $Y$, и известна их ковариационная матрица:

$Var\dbinom XY=\begin{pmatrix}
100 & -9\\-9 & 81
\end{pmatrix}$

Требуется найти:

а) на сколько в среднем растёт $Y$ при росте $X$ на 1?

б) на сколько в среднем растёт $Y$ при росте $X$ на 1?

в) $Corr(X,Y)$

**Отступление 1: Что видно сразу?**

1) Так как ковариационная матрица составлена из попарных ковариаций случайных величин (подробнее в [Конспекте 8](https://github.com/bdemeshev/pr201/blob/master/2016/stud_notes/08_pr201_notes.ipynb)), то 

$Cov(X,X)=Var(X)=100$

$Cov(Y,Y)=Var(Y)=81$

$Cov(X,Y)=Cov(Y,X)=-9$

2) $X$ и $Y$ зависимы

А теперь решение!

**а)** Пусть $Y=\beta X + \text{хвост}_Y$, где 

* $Y$ — случайная величина
* $\beta X$ — случайная величина, линейно связанная с $Y$
* $\text{хвост}_Y$ — случайная величина, линейно не связанная с $X$
* $\beta$ — константа

Поскольку $\text{хвост}_Y$ и $X$ независимы, $Cov(X,\text{хвост}_Y)=0$. Далее, выразим $\text{хвост}_Y$ и найдём $\beta$:

$Cov(X,Y-\beta X)=0$

$Cov(X,Y)-Cov(X,\beta X)=Cov(X,Y)-\beta Var(X)=-9-\beta \cdot 100=0 \Rightarrow \boxed{\beta = \text{–}0.09}$

_Вывод:_ с ростом $X$ на $1$ $Y$ будет в срднем уменьшаться на $0.09$

**б)** Пусть $X=\gamma Y + \text{хвост}_X$

Заметим, что в предыдущем пункте коэффициент $\beta = \dfrac{Cov(X,Y)}{Var(X)}$. Тогда аналогично, $\gamma = \dfrac{Cov(Y,X)}{Var(Y)}$, откуда получим:

$\gamma = \dfrac{Cov(Y,X)}{Var(Y)} = \dfrac{-9}{81} \Rightarrow \boxed{\gamma= - \dfrac{1}{9} }$

_Вывод:_ с ростом $Y$ на $1$ $X$ будет в срднем уменьшаться на $\dfrac{1}{9}$


**Отступление 2: Почему $\gamma$ не равно $\beta$ ?**

Вернёмся в довероятностную эпоху и рассмотрим детерминистическую функцию $y=f(x)$

Как изветсно из математического анализа, для взаимно обратных функций $f: X \to Y$ и $f^{-1}: Y \to X$, в точках $x_0 \in X$ и $f(x_0)=y_0 \in Y$ выполняется равенство: $(f^{-1})'(y_0) = (f'(x_0))^{-1}$, а геометрически: касательная к обратной функции будет иметь обратный наклон.

Но теперь у нас появилась случайная величина $\text{хвост}_Y$, и если выразить $X$ из пункта а) как $X=\dfrac{1}{\beta}Y-\dfrac{1}{\beta}\text{хвост}_Y$ и изменить первое слагаемое, изменяя $Y$, то поменяется и второе слагаемое, так как случайная величина $\text{хвост}_Y$ связана со случайной величиной $Y$ (вот так: $Var(Y) = \beta ^2 Var(X)+Var(\text{хвост}_Y)$, что и можно увидеть на рисунке). Поэтому $\beta \neq \dfrac{1}{\gamma}$: при изменениях $Y$ с.в. $X$ меняется за счёт обоих слагаемых!

![image1](http://s020.radikal.ru/i720/1611/81/34e13ca1ebe2.png)

**в)** Кульминация!

Перемножаем полученные в пунктах а) и б) выражения для $\beta$ и $\gamma$:

$\beta \cdot \gamma = \dfrac{Cov(X,Y)^2}{Var(x)Var(Y)}$ — и видим знакомую формулу справа :)

$Corr(X,Y)=\dfrac{Cov(X,Y)}{\sigma_X \sigma_y}= \dfrac{-9}{\sqrt{81\cdot 100}}=-\dfrac{1}{10}$ или $Corr(X,Y)=-\sqrt{\beta \cdot \gamma} = -\sqrt{\dfrac{-9}{100}\cdot \dfrac{-1}{9}}$

&#128073; _Очень важный вывод (aka тайный смысл корреляции):_ корреляция — это среднее геометрическое между силой, с которой $X$ меняет $Y$, и силой, с которой $Y$ меняет $X$ &#128072;

# Задача 1 из листка "Совместная плотность"

**Условие:** на первом шаге значение $X$ выбирается случайно равномерно на отрезке $[0,1]$, на втором шаге значение $Y$ выбирается случайно и равномерно на от $0$ до полученного $X$. Найти функции плотности: $f(y|x), f(x), f(x,y), f(x|y), f(y)$

**Решение:**

* $f(y|x)$ 

Фиксируем $x$ и смотрим, как при этом меняется $y$, и видим, что $y$ равномерно распределён на отрезке $[0, x]$:

$
f(y|x) =
    \begin{cases}
        \dfrac{1}{x-0}, & \text{если } y \in (0,x) \\
        0, & \text{иначе}
    \end{cases}
$

![image2](http://s017.radikal.ru/i440/1611/b5/e28b595dfea0.png)


* $f(x)$ 

Из условия: $X$ равномерно распределена на $[0,1]$, поэтому:

$
f(x) =
    \begin{cases}
        1, & \text{если } x \in [0,1] \\
        0, & \text{иначе}
    \end{cases}
$

* $f(x,y)$

Время вспомнить про условные вероятности!

$P(A|B)=\dfrac{P(A \cap B)}{P (B)}$

C функцией плотности поступаем аналогично:

$f(y|x)=\dfrac{f(x,y)}{f(x)}$

**Отступление 3: Про запись**

Формально писать $\text{так } f(y|x)$ нельзя!

А как надо писать? Вот так: $f_{Y|X}(x,y)$

А кто так пишет? Никто!

Вернёмся к решению. Находим $f(x,y)=f(y|x)\cdot f(x)$,тогда получим:

$
f(x,y) =
    \begin{cases}
        \dfrac{1}{x}, & \text{если } x \in (0,1] \text{ и } y \in [0,x) \\
        0, & \text{иначе}
    \end{cases}
$

* $f(y)$

$f(y) = \int_{\text{-}\infty}^{\infty} f(x,y) dx$    


Определяем пределы интегрирования. Для этого обратимся к рисунку: треугольник — это область, полученная нами из условия о том, что $X$ принимает значения из промежутка $[0,1]$, а $Y$ — от $0$ до $X$; над этой областью лежит их своместная функция плтоности $f(x,y)=\dfrac{1}{x}$, вне её функция принимает значение $0$. Далее, мы фиксируем какоое-нибудь значение на оси $0Y$ и заставляем $x$ бегать (дистанция, которую пробегает $x$ обозначена горизонтальными линиями из точек). На рисунке видно, что можно выбрать такое значение на оси $0Y$, что ни один $x$ не попадёт в треугольник, но если взять такое значение, что $x$ всё-таки добежит до треугольника, то получим, что $x$ пробегает отрезок от $y$ до $1$ — это и есть пределы интегрирования!



![image3](http://s019.radikal.ru/i600/1611/dc/b55f79228a30.png)

Считаем:

$$f(y) = \int_{\text{-}\infty}^{\infty} f(x,y) dx =\int_{y}^{1} \dfrac{1}{x} dx = -\text{ln } y$$

Для тех, кого смутил знак "минус" перед логарифмом, напоминаю, что $0 \le y \le 1$ , так что всё ок!

А вот и полный ответ:

$
f(y) =
    \begin{cases}
        -\text{ln } y , & \text{если } y \in (0,x) \\
        0, & \text{иначе}
    \end{cases}
$

* $f(x|y)$

Снова используем соотношение $f(x|y)=\dfrac{f(x,y)}{f(y)}$ и получаем:

$
f(x|y) =
    \begin{cases}
        -\dfrac{1}{x\text{ ln } y }, & \text{если }x \in (0,1] \text{ и } y \in (0,x) \\
        0, & \text{иначе}
    \end{cases}
$
