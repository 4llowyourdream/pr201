## Упражнение 20.5
Вася наугад поймал 400 покемонов-девочек и 100 покемонов-мальчиков. Средний
рост покемонов-девочек равен 0.9 метра, и сумма квадратов ростов равна $\sum\limits_{i = 1}^n Y^2_{i} = 1000$. i
Для покемонов мальчиков средний рост равен 1 метру, а сумма квадратов ростов равна $\sum\limits_{i = 1}^n X^2_{i} = 2000$.
1. Постройте точечную оценку для ожидания роста покемона-мальчика $\mu_{X}$.
2. Постройте точечную оценку для дисперсии роста покемона-мальчика $\sigma^2_{X}$.
3. Постройте 95%-ый доверительный интервал для ожидаемого роста покемона-мальчика.
4. Постройте 95%-ый доверительный интервал для ожидаемого роста покемона-девочки.
5. Постройте 95%-ый доверительный интервал для разницы ожидаемых ростов покемона-мальчика и покемона-девочки.
6. Нужно ли предположение о нормальности $X_{i}$ и $Y_{i}$ для решения предыдущих пунктов?


## Решение:
    
1) Так как имеется 100 наблюдений для мальчиков покемонов, то из ЦПТ следует, что  $\sum\limits_{i = 1}^{100}
X_{i} \sim N$ 

Используя метод моментов получаем
$\hat{\mu}_{X} = \bar{X} = 1$

2)$\hat{\sigma}^2_{X} = \sum\limits_{i = 1}^{n} \frac{(X_{i} - \bar{X})^2}{n-1} = \frac{\sum\limits_{i = 1}^{n} (X^2_{i}) - 2\bar{X}\sum\limits_{i = 1}^{n}X_{i}+n\bar{X}^2)}{n-1} = \frac{\sum\limits_{i = 1}^{n} (X^2_{i}) - n\bar{X}}{n-1} = \frac{1900}{99}$ 

В данном уравнении мы использовали факт, что $\sum\limits_{i = 1}^{n}X_{i} = n\bar{X} $

Уравнение 
$\sum\limits_{i = 1}^{n} (X_{i} - \bar{X})^2= \sum\limits_{i = 1}^{n} (X^2_{i}) - n\bar{X}$ можно доказать и  с помощью теоремы Пифагора, если взять за катеты векторы $\bar{X} = (\bar{X},\bar{X},..,\bar{X}), Y = (X_{1}-\bar{X},X_{2}-\bar{X},..,X_{n}-\bar{X})$, а за гипотенузу вектор $X = (X_{1},X_{2},..,X_{n}) $

Заметим, что $Y+\bar{X} = X $ 

$<\bar{X},Y> = \sum\limits_{i = 1}^{n}\bar{X}\cdot(X_{i}-\bar{X})= \bar{X} \cdot\sum\limits_{i = 1}^{n}(X_{i}-\bar{X}) = 0$ 

Следовательно $\bar{X}$ и $Y$ перпендикулярны, $=> |X^{2}| = |\bar{X}^{2}| + |Y^{2}| => \sum\limits_{i = 1}^{n}(X_{i}-\bar{X})^{2} = \sum\limits_{i = 1}^{n}(X_{i})^{2} - n \cdot \bar{X}^{2}$



3)По ЦПТ имеем, что $\bar{X} \sim N({\mu},\frac{\sigma^2}{n})$,

следовательно:

$\frac{\bar{X} - \mu}{\sqrt{\frac{\sigma^2}{n}}} \sim N(1,0)$

$P(\frac{\bar{X} - \mu}{\sqrt{\frac{\sigma^2}{n}}}< z_{\frac{\alpha}{2}}) = 0.95$

Сделав преобразования, получаем, стандартный доверительный интервал для математического ожидания
$ \bar{X} - z_{\frac{\alpha}{2}}\sqrt{\frac{\sigma^2}{n}} < \mu < \bar{X} + z_{\frac{\alpha}{2}}\sqrt{\frac{\sigma^2}{n}}$

Вместо теоритической дисперсии подставим выборочную дисперсию и получим 

$ \bar{X} - z_{\frac{\alpha}{2}}\sqrt{\frac{\hat{\sigma}^2}{n}} < \mu < \bar{X} + z_{\frac{\alpha}{2}}\sqrt{\frac{\hat{\sigma}^2}{n}} (\cdot)$

где $z_{\frac{\alpha}{2}} = 1.96$ (Получаем из таблицы нормального распределения)


Подставляя все известные значения получаем доверительный интервал

$ 0.142 < \mu < 1.858 $ 

4)Сделаем все действия и для покемонов девочек, получим

$\hat{\mu}_{Y} = \bar{Y} = 0.9$

$\hat{\sigma}^2_{Y} = \frac{\sum\limits_{i = 1}^{n} (Y^2_{i}) - n\bar{Y}}{n-1} = \frac{640}{399}$ 

Осталось подставить $\hat{\sigma}^2_{Y}$  и  $\bar{Y}$ в формулу доверительного интервала $\cdot$ и получить 95% интервал ожидаемого роста для девочек-покемонов
$0.876<{\mu}_{Y}<1.024$

5)Мы знаем, что рост покемонов девочек и рост покемонов мальчиков независимые величины, следовательно можно утверждать что.

1.$\bar{Х}- \bar{Y} = \hat{\mu}_{X} - \hat{\mu}_{Y} = 0.1$

2.$Var(\bar{Х}- \bar{Y}) = \frac{\sigma_{X}^2}{n}+\frac{\sigma_{y}^2}{n}$

из 1. и 2. следует, что $\bar{Х}- \bar{Y} \sim N(\hat{\mu}_{X} - \hat{\mu}_{Y},\frac{\sigma_{X}^2}{n}+\frac{\sigma_{y}^2}{n})$


По ЦПТ можно утверждать, что

$\frac{\bar{Х}- \bar{Y} - (\hat{\mu}_{X} - \hat{\mu}_{Y})}{\sqrt{\frac{\sigma_{X}^2}{n}+\frac{\sigma_{y}^2}{n}}}\sim N(0,1)$


Осталось построить доверительный интервал

$P(z_{\frac{-\alpha}{2}}<\frac{\bar{Х}- \bar{Y} - ({\mu}_{X} - {\mu}_{Y})}{\sqrt{\frac{\sigma_{X}^2}{n}+\frac{\sigma_{y}^2}{n}}}<z_{\frac{\alpha}{2}}) = 0.95$

В данном случае, тоже, берем оценку дисперсии

$P(\bar{Х}- \bar{Y}-z_{\frac{-\alpha}{2}}\sqrt{\frac{\hat{\sigma_{X}}^2}{n}+\frac{\hat{\sigma_{y}}^2}{n}} < {\mu}_{X} - {\mu}_{Y}<\bar{Х}- \bar{Y}+z_{\frac{\alpha}{2}}\sqrt{\frac{\hat{\sigma_{X}}^2}{n}+\frac{\hat{\sigma_{y}}^2}{n}})= 0.95$    ,где  $z_{\frac{\alpha}{2}} = 1.96 $


Подставляя все значения, получаем итоговый 95% доверительный интервал для разницы ростов между мальчиками и девочками покемонами

$-0.7675<{\mu}_{X} - {\mu}_{Y}<0.9675$
