## №1

Пусть А — событие, когда Петя выбирает неправильную монетку;
В — событие, когда выбирает правильную и два раза выпадает орел

Построим дерево вариантов:


![](https://www.draw.io/?libs=general;flowchart#WD1982B9D6CD3AE19!394)

Отсюда видно, что Р(Выбор неправильной монеты — Орёл — Орёл)=$\frac{1}{3}*1*1=\frac{1}{3}$, а Р(В)=$\frac{2}{3}*\frac{1}{2}*\frac{1}{2}=\frac{2}{12}$.

Теперь просто посчитаем вероятность события, что монета "неправильная" и при этом выпадает 2 раза орёл:

$$
P(A|B)=\dfrac{P(A \cap B)}{P(B)} 
=\dfrac{\frac{1}{3}}{\frac{2}{12}+\frac{1}{3}}
=\dfrac{\frac{1}{3}}{\frac{6}{12}}
=\dfrac{2}{3}
$$

## №2

Пусть А — событие, когда попал 1-ый охотник; В — событие, когда утка убита одним выстрелом.

![](https://www.draw.io/?libs=general;flowchart#WD1982B9D6CD3AE19!395)

Из дерева вариантов видно, что Р(Первый попал — 1 выстрел) = $0.3 * 0.4$, а Р(В) = $(0.3 * 0.4)+(0.7 * 0.6) $ $\Rightarrow$



$$
P(A|B)=\dfrac{P(A \cap B)}{P(B)} 
=\dfrac{0,3 * 0,4}{(0,3 * 0,4)+(0,7 * 0,6)}
=\dfrac{0,12}{0,12+0,42}
=\dfrac{0,12}{0,54}=\frac{2}{9}
$$

## №4

Пусть А — событие, отражающее, что у Пети есть два туза как минимум, В — событие, когда у него точно есть 1 туз.
Для удобства примем X  — количество тузов, а также воспользуемся формулой $$C^k_n = \dfrac{n!}{k!(n-k)!}$$

Для начала найдем вероятность наличия у Пети хотя бы одного туза (у нас есть сумма вероятностей всех возможных событий, равная 1. Если из нее вычесть вероятность того, что у Пети точно нет ни одного туза, то результатом будет как раз искомая вероятность)

$$P(B)=P(X\geqslant 1)=1 - P(X=0)=1 - \dfrac{C^{13}_{48}}{C^{13}_{52}}=$$ $$= 1 - \dfrac{\frac{48!}{13!(48-13)!}}{\frac{52!}{13!(52-13)!}}= 1 - \dfrac{36*37*38*39}{49*50*51*52}=0,6961824729891957$$


(Поступим аналогично случаю нахождения Р(В))
$$ P(X=0) = \dfrac{C^{13}_{48}}{C^{13}_{52}}= \dfrac{\frac{48!}{13!(48-13)!}}{\frac{52!}{13!(52-13)!}}= \dfrac{36*37*38*39}{49*50*51*52}= 0,3038175270108043$$


$$ P(X=1) =  \dfrac{С^{1}_{4} * С^{12}_{48}}{С^{13}_{52}}=\dfrac{\frac{4!}{1!(4-1)!} * \frac{48!}{12!(48-12)!}}{\frac{52!}{13!(52-13)!}}=0,4388475390156062$$

$$ P(X\geqslant 2)= 1 - P(X=1) - P(X=0) = 0,2573349339735895$$

$$ P(X\geqslant 2 | X\geqslant 1) = \dfrac{P(X\geqslant 2 \cap X\geqslant 1}{X\geqslant 1}) = \dfrac{P(X\geqslant 2)}{P(X\geqslant 1)} = 0,3696371913367362$$
