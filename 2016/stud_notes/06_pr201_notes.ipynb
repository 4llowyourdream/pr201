## Задача № 7(из контрольной прошлого года)

Нам дано время, за которое студенты выполняют задание - случайная величина с функцией плотности 


\begin{equation*}
f(x) =  \begin{cases}
3x^2,\ если\ x \in [0;1]\\
0,\ иначе
  \end{cases}
 \end{equation*}


 Необходимо найти:
 
 (1) Функцию распределения случайной величины $X$
 
 (2) $P(случайно\ выбранный\ студент\ закончит\ менее\ чем\ за\ \frac{1}{2}\ часа$
 
 (3) Медиана $X$
 
 (4) Функцию распределения случайной величины $Y=\frac{1}{X}$
 
 (5)$P\left(студент,\ закончивший\ не\ менее\ чем\ за\ 15\ минут,\ закончит\ не\ более\ чем\ за\ 30\right)$
 
$\textit{Решение:}$
 
 1)По определению - функция распределения в точке $x$ - вероятность того, что случайная величина $X<x$. Вероятность равна интегралу функции плотности до заданной точки:

F(x)= \begin{cases}
0,\ если\ x<0\\
x^3, если\ x \in [0;1]\\
1, если\ x>1
\end{cases}
\end{equation*}

2) Так как $\frac{1}{2}$ попадает в отрезок $[0;1]$, то по выведенному в предыдущем пункте:

\begin{equation*}
P\left(X<\frac{1}{2}\right)=\int_{0}^{\frac{1}{2}} 3x^2dx=
x^3 \Bigr|^\frac{1}{2}_0=\frac{1}{8}-0=\frac{1}{8}
\end{equation*}

3)Найдем медиану распределения $m$: в нашем случае (недискретном) медиана - такое $\textit{число}$, что вероятность попасть правее равна вероятности попасть левее и равна $\frac{1}{2}$. Но вероятность попасть правее $= \int_{0}^{m}3x^2dx = \frac{1}{2}$

\begin{equation*}
x^3 \Bigr|^\frac{1}{2}_0=\frac{1}{2}\\
x^3=\frac{1}{2}\\
x=\frac{1}{\sqrt[3]{2}} - медиана\ распределения.
\end{equation*}

4)$F_Y(t)=P(Y\le t)=P\left(\frac{1}{X}\le t\right)=P\left(X\ge \frac{1}{t}\right)$

$X \in [0;1]$

$t=\frac{1}{X}\Rightarrow t \in [1;+\infty)$

Если $t$ не попадает в этот диапазон


$t\le 1$

Тогда $P(\frac{1}{X}\le t)=P\left(\frac{1}{X}\le\frac{1}{2}\right)=0, так\ как\ \frac{1}{X} \in [1;+\infty)$

Теперь предположим $t>1$. Тогда применим обратную функцию:

$P\left(X\ge \frac{1}{t}\right)=1-P\left(X\le \frac{1}{t}\right)=1-F_X\left(\frac{1}{t}\right)=1-\frac{1}{t^3}$

Таким образом:

\begin{equation*}
F_Y(t)= \begin{cases}
1-\frac{1}{t^3},\ если\ t\ge 1\\
0,\ если\ t<1
\end{cases}
\end{equation*}

Рассчитаем для $Y=2$:

$P\left(Y\le 2\right)=1-\frac{1}{2^3}=\frac{7}{8}$

Кроме того, мы можем найти функцию плотности величины $Y$ как производную от функции распределения:

$F'_Y(t)=\frac{3}{t^4}$

Тогда:

\begin{equation*}
f_Y(t)= \begin{cases}
\frac{3}{t^4},\ если\ t\ge 1\\
0,\ если\ t<1
\end{cases}
\end{equation*}

5)По формуле условной вероятности:

$P\left(\left.X>\frac{1}{2}\right|X>\frac{1}{4}\right)=\frac{P\left(X>\frac{1}{2}\right)}{P\left(X>\frac{1}{4}\right)}=\frac{1-\frac{1}{8}}{1-\frac{1}{64}}=\frac{7\cdot 64}{8\cdot 63}=\frac{8}{9}$





## Задача про диктатора(очень странного диктатора)

<a href="http://vfl.ru/fotos/bf09a10914640816.html"><img src="//images.vfl.ru/ii/1477271640/bf09a109/14640816_m.jpg" alt="Диктатор" title="Диктатор" border="0"></a>


Условие: пусть в Некоторой стране живет $N>>0$ семей, каждая из которых производит на свет детей. По приказу диктатора, как только в i-той семье появляется первая девочка, эта семья останавливает "производство". Чему равна ожидаемая доля мальчиков в стране от введения такой меры? А чему равна ожидаемая доля мальчиков в случайно выбранной семье?

Решение: сначала интуитивно кажется, что ответы должны быть одинаковы. Однако не всегда стоит полагаться на интуицию. И вот почему:

Понятно, что в каждой семье будет ровно одна девочка. Рассмотрим, с какой вероятностью встречаются различные конфигурации:

$P(D)=\frac{1}{2};\ R=0(R - доля\ мальчиков\ в\ такой\ конфигурации)$
$P(DM)=\frac{1}{4};\ R=\frac{1}{2}$

$P(DMM)=\frac{1}{8};\ R=\frac{2}{3}$

Наблюдаем закономерность, из которой выводим:

$E(R)=\sum_{k=1}^\infty \frac{k-1}{k}\cdot 2^{-k}=\sum_{k=1}^\infty 2^{-k} - \sum_{k=1}^\infty \frac{2^{-k}}{k}$

Второе слагаемое посчитать проблематично, поэтому применим хитрость: заменим $\frac{1}{2}$ на $a$. И тут же продифферинцируем:


$F(a)=\sum_{k=1}^\infty \frac{a^k}{k}$

$F'(a)=\sum_{k=1}^\infty a^{k-1}=\frac{1}{1-a}$ (это же бесконечно убывающая геометрическая прогрессия!)

А теперь проинтегрируем обратно:

$F(a)=-ln(1-a)+c$

У нас $a=\frac{1}{2}$, значит, $E(R)=1-ln2$ - ожидаемая доля мальчиков в случайно выбранной семье.





Теперь посчитаем ожидаемую долю мальчиков во всей стране. Обозначим за $C_i$ количество детей в i-той семье. Тогда:

$E\left(\frac{R_1C_1+R_2C_2+\cdots +R_NC_N}{C_1+C_2+\cdots +C_N}\right)=E\left(\frac{C_1-1+C_2-1+\cdots +C_N-1}{C_1+C_2+\cdots +C_N}\right)=E\left(1-\frac{n}{\sum C_i}\right)$– вот чему будет равно ожидаемая доля мальчиков в стране.

Рассмотрим выражение $E\left(\frac{\sum C_i}{N}\right)$.
Оно равно $E\left(C_i\right)(просто\ надо\ сократить\ на\ N)$

Можно составить уравнение относительно $E\left(C_i\right)$

Как только семья решает произвести первого ребенка, с вероятностью $\frac{1}{2}$ появляется девочка, и с вероятностью $\frac{1}{2}$ мы возвращаемся в исходную точку, прибавляя еще один шаг. Тогда:

$E\left(C_i\right)=\frac{1}{2}\cdot 1+ \frac{1}{2}\cdot \left(E\left(C_i\right)+1\right)\Longrightarrow C_i=\frac{1}{2}$

Следовательно, $\frac{\sum C_i}{N}\to 2\Longrightarrow 1-\frac{N}{\sum C_i}\to \frac{1}{2}$

Интересный вывод: ожидаемая доля мальчиков в семье не равна ожидаемой доли мальчиков в стране. Объясняется достаточно просто: количество детей в семье может быть совершенно разным. Если в семье очень много мальчиков и 1 девочка, то эта семья дает большой вклад при подсчтете доли мальчиков в стране, однако такой же как и все вклад при подсчете доли мальчиков в случайно выбранной семье. Поэтому ожидаемая доля мальчиков в стране больше, чем в случайной семье.
