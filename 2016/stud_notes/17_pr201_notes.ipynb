#### Задача 15.1

*Имеется пять чисел: $\mathcal{x}$, 9, 5, 4, 7. При каком значении $\mathcal{x}$ медиана будет равна среднему?*

Расставим числа в порядке возрастания. 

$\mathcal{x}$ 4 $\mathcal{x}$ 5 $\mathcal{x}$ 7 $\mathcal{x}$ 9 $\mathcal{x}$

В зависимости от расположения $\mathcal{x}$ меняется медиана: 5, $\mathcal{x}$ или 7. Составим 3 уравнения:

$\frac{\mathcal{x}+4+5+7+9}{5}$=5
$\Rightarrow$ $\mathcal{x}$=0

$\frac{\mathcal{x}+4+5+7+9}{5}$=$\mathcal{x}$
$\Rightarrow$ $\mathcal{x}$=6,25

$\frac{\mathcal{x}+4+5+7+9}{5}$=7
$\Rightarrow$ $\mathcal{x}$=10

Подумать: А можно ли поставить такие числа в условии задачи, чтоб $\mathcal{x}$ не существовал?

#### Задача 15.2

*Измерен рост 25 человек. Средний рост оказался равным 160 см. Медиана оказалась равной 155 см. Машин рост в 163 см был ошибочно внесен как 173 см. Как изменится медиана и среднее после исправления ошибки? А как могут измениться медиана и среднее, если рост Маши равен 153?*

Если рост Маши ошибочно был внесен как 173 см вместо 163, то при исправлении ошибки изменения никак не отразятся на медиане, потому что ошибочно внесенный рост и ее рост больше медианы. Средний рост уменьшится.

#### Задача 15.4

*Деканат утверждает, что если студента N перевести из группы А в группу В,то средний рейтинг каждой группы возрастет. Возможно ли такое?*

Да, возможно. Если средняя оценка N ниже средней оценки группы А, но выше средней в группе В, то после смены студентом группы средняя оценка каждой группы и ее рейтинг возрастут.

Например, группы А и В состоят из 3 человек, которые имеют оценки 8, 9, 10 и 1, 2, 3 соответственно. Студент N, имеющий гценку 8, желает перейти в группу В. Тогда изменения оценки группы А = $\frac{9 + 10}{2}$ - $\frac{9 + 10 + 8}{3}$ = 0,5, то есть рейтинг группы повысится; изменения для группы В: $\frac{1 + 2 + 3 + 8}{4}$ - $\frac{1 + 2 + 3}{3}$ = 1,5, а значит рейтинг группы В тоже повысится.

#### Задача 15.8

*Во время Второй Мировой войны американские военнные собрали статистику попаданий пуль в фюзеляж самолёта. По самолётам, вернувшимся из полёта на базу, была составлена карта повреждений среднестатистического самолёта. С этими данными военные обратились к статистику Абрахаму Вальду с вопросом, в каких местах следует увеличить броню самолёта. Что посоветовал Абрахам Вальд и почему?*


![](http://localhost:8888/files/Survivorship-bias.png)

Абрахам Вальд посоветовал укреплять те места, где было меньше всего повреждений. Статист располагал данными только по вернувшимся самолетам. Места, в которых было наибольшее количество повреждений, были не так опасны, потому что позволяли самолетам вернуться. Места с наименьшими повреждениями были менее защищенными и чаще приводили к тому, что самолет сбивали, а поэтому они требовали укрепления. 

Через несколько лет Абрахам Вальд написал $\href{http://www4.ncsu.edu/~swu6/documents/A_Reprint_Plane_Vulnerability.pdf}{работу}$ "A method of estimating plane vulnerability based on damage of survivors", которая посвящена этим событиям. Она очень интересна!
