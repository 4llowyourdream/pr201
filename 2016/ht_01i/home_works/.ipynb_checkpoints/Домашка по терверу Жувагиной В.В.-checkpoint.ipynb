```python
a = stat.uniform(loc=0,scale=1)
```

```python
b = a.rvs(1000000)
```

```python
sns.plt.title('Для X1')
sns.distplot(b, axlabel = 'X1')
```

```python
c = a.rvs(1000000)
d = a.rvs(1000000)
```

```python
sns.plt.title('Для X1 + X2 + X3')
sns.distplot(b + c + d, axlabel = 'X1 + X2 + X3')
```

```python
summ = np.empty(shape=100000)
for i in range(10):
    v = a.rvs(100000)
    summ += v

sns.plt.title('Для $\sum_{i = 1}^{10} X_i$')
sns.distplot(summ)
```

```python
summm = np.empty(shape = 100000)
for i in range(50):
    v = a.rvs(100000)
    summm += v
sns.plt.title('Для $\sum_{i = 1}^{50} X_i$', y = 1.02)
sns.distplot(summm)
```

## Задача 3 (№ 20)

У тети Маши - двое детей разных возрастов. Вероятности рождения девочки и мальчика равны.

### Пункт 1

Какова вероятность того, что у тёти Маши оба ребёнка-мальчики, если известно, что у неё хотя бы один ребёнок-мальчик?

P(B)- вероятность рождения мальчика 
P(G)-вероятность рождения девочки. 
Возможные исходы: BB, GG, BG, GB.
 Нужно найти условную вероятность:  $P(\text{BB | хотя бы один B}) = \dfrac{P(BB \cap \text{хотя бы один B)}}{P(\text{хотя бы один B})} $

$P(\text{BB | хотя бы один B)} =  \dfrac {1/4}{3/4} = \dfrac{1}{3}$

Ответ: $\dfrac{1}{3}$

### Пункт 2

Какова вероятность того, что у тёти Маши оба ребёнка - мальчики, если известно, что у неё хотя бы один ребёнок - мальчик водолей по гороскопу?

 Нужно найти условную вероятность:  $P(\text{BB | хотя бы один B водолей}) = \dfrac{P(BB \cap \text{хотя бы один B водолей)}}{P(\text{хотя бы один B водолей})} $

$P(\text{BB | хотя бы один B водолей)} =  \dfrac {1/48 + 11/48}{1/24 + 11/48 * 1/12 + 1/48} = \dfrac{4}{63}$

Ответ:  $\dfrac{4}{63}$
