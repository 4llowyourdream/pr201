Так как у меня очень плохие оценки, а я хочу высокую оценку за курс, то решим задачу для произвольного количества голосов. Пусть количество голосов за кандидата $А$ равняется m а за кандидата $Б – n$. Количество исходов, очевидно, равняется количеству перестановок $C_m^{n+m}$ . Пусть  $Q_{n+m}$ – количество неблагоприятных исходов. Такие перестановки начинаются или с $А$ или с $Б$. 
Поставим в соответствие i-тому голосу букву $А$ или $Б$ в распределении в зависимости от того за кого был отдан голос. Количество неблагоприятных исходов, начинающихся с $Б$, равняется количеству всевозможных перестановок образованных из n голосов для $А$ и $m-1$ голосов для $Б$. Потому что, очевидно, достаточно зафиксировать первую букву $Б$ и мы получим все остальные варианты.
Количество неблагоприятных исходов, начинающихся с $А$ также представимо в приведенной выше форме. Потому что существует простое правило, с помощью которого можно составить прямое соответствие с перестановкой образованной из n голосов для $А$ и $m-1$ голосов для $Б$.


Правило состоит из двух частей:
	

   1)В  неблагоприятной перестановке, начинающейся с $А$, нужно удалить первый встречающийся голос за $Б$, который нарушает правило(уравнивает количество голосов за $А$ и $Б$. Затем нужно поменять местами части перестановки, которые были разделены удаленной буквой. Тогда мы получим определенную единственным образом перестановку из $n$ голосов для $А$ и $m-1$ голосов для $Б$. Положим, например, неблагоприятная последовательность голосов $ААББАБАА$, она состоит из 5 голосов за $А$ и 3 голосов за $Б$. Удаляя первую букву $Б$, нарушающую наш закон, мы получаем две отдельные группы $ААБ$ и $АБАА$. По вышеописанному правилу получаем перестановку $АБААААБ$, состоящую из 5 голосов за $А$ и 2 за $Б$.
	

2)Пусть дана произвольная перестановка из $n$ голосов за $А$ и $m-1$ голосов за $Б$. Мы проходим справа на лево до тех пор, пока  мы не получим группу, где количество $А$ превышает (на один) количество $Б$.Рассмотрим эту группу и буквы, расположенные левее. Дальше меняем эти группы местами, и ставим между ними букву $Б$. Таким образом получаем неблагоприятную перестановку, начинающуюся с $A$ и исключительно уникальную. Возьмем вышеприведенный пример. Получим две группы $ААБ$ и $АБАА$, затем поменяв их местами и вставив букву $Б$ получаем исходную перестановку.



Из всего вышеупомянутого следует,  что общее количество неблагоприятных результатов

$Q_{n+m}=2\frac{(n+m-1)!}{n!(m-1)!}$


Таким образом, количество благоприятных событий равно:

А соответствующая вероятность равна:
$\frac{\frac{(n+m-1)!}{n!m!} (n-m)}{ \frac{(n+m)!}{n!m!}}=\frac{n-m}{n+m}$
