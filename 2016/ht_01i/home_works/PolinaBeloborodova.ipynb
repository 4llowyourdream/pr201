```python
url = 'https://raw.githubusercontent.com/johan/world.geo.json/master/countries.geo.json' 

r = requests.get(url) 
geo_json_data = r.json() 

def get_coordinates(features): 
     depth = lambda L: isinstance(L, list) and max(map(depth, L))+1 
     xs = [] 
     ys = [] 
     for feature in geo_json_data['features']: 
         coords = feature['geometry']['coordinates'] 
         nbdims = depth(coords) 
         # one border 
         if nbdims == 3: 
             pts = np.array(coords[0], 'f') 
             xs.append(pts[:, 0]) 
             ys.append(pts[:, 1]) 
         # several borders 
         else: 
             for shape in coords: 
                 pts = np.array(shape[0], 'f') 
                 xs.append(pts[:, 0]) 
                 ys.append(pts[:, 1]) 
     return xs, ys 


```

```python
TOOLS = "pan,wheel_zoom,box_zoom,reset,hover,save"

xs, ys = get_coordinates(geo_json_data['features']) 
p = figure(plot_width=900, plot_height=600, x_range=(-180,180), 
y_range=(-90,90), tools=TOOLS) 

p.patches(xs, ys, fill_color="#F1EEF6", fill_alpha=0.7, line_width=2) 
```

```python
list_params = ['movie_facebook_likes', 'title_year', 'imdb_score', 'budget', 'gross']
```

```python
def distribution(params):
    
    def coaff(params):
        if params == 'gross':
            return(2500)
        if params == 'imdb_score':
            return(320)
        if params == 'budget':
            return(2500)
        if params == 'title_year':
            return(3600)
        else:
            return(1800)
    coaff = coaff(params)
    p1 = figure(title="Distribution of "+params,tools="save",
                background_fill_color="#E8DDCB")

    hist, edges = np.histogram(data[params], density=True, bins=50)

    p1.quad(top=hist*coaff, bottom=0, left=edges[:-1], right=edges[1:],
            fill_color="#036567", line_color="#033649")

    p1.xaxis.axis_label = params

    p1.line(data.groupby(params)['movie_title'].count().keys(), data.groupby(params)['movie_title'].count(), line_width=5, line_color="#040564", alpha=0.5)

    show(p1)
```

```python
interact(distribution, params = list_params)
```

# Задание 3

Рассмотрим все варианты скрещивания и вероятности:

$AA*AA - p_1^2$

$AA*Aa - 2*p_1*p_2$

$AA*aa - 2*p_1*p_3$

$Aa*Aa - p_2^2$

$Aa*aa - 2*p_2*p_3$

$aa*aa - p_3^3$

Зная, какие дети рождаются у всевозможных пар, составим аналогичную таблицу вероятностей для генотипов детей (свернув далее в квадраты):

$AA - p_1^2+p_1*p_2+0.25*p_2^2 = p^2$

$Aa - p_1*p_2+p_2*p_3+0.5*p_2^2+2*p_1*p_3 = 2pq$

$aa - p_3^2+p_2*p_3+0.25*p_2^2 = q^2$



Видим, что все это сворачивается в полные квадраты:

$AA - (p_1+0.5*p_2)^2 = p'^2$

$Aa - 2(p_1+0.5*p_2)^2*(p_3+0.5*p_2)^2 = 2p'q'$

$aa - (p_3+0.5*p_2)^2 = q'^2$


Соотношение генов не изменилось и все еще удовлетворяет уравнению $p'^2+2p'q'+q'^2=1$

Потому соотношение генов в любом поколении не меняется, а значит, и утверждение, что рецессивный ген со временем исчезает - неправда.
