### Пункт "1"

Пусть есть N независимых равномерно распределённых случайных величин X. 

Построим гистограмму равномерного распределения $X_{1} \sim U(0,1)$, функция плотности которого:

$$ f(x) = \left\{
\begin{array}{rcl}
1, &x \geq 0 \\
0, & x \in (-\infty; 0) \cup (1; +\infty)  \\ 
\end{array}
\right.
$$

```python
X1 = stats.uniform(0,1)
values1 = X1.rvs(10000000)
sns.distplot(values1)
```

Построим гистограмму распределения для суммы случайных величин: $X_{1}+X_{2}+X_{3}$

```python
X1 = stats.uniform(0,1)
values1 = X1.rvs(10000000)
X2 = stats.uniform(0,1)
values2 = X2.rvs(10000000)
X3 = stats.uniform(0,1)
values3 = X3.rvs(10000000)
Value = values1 + values2 + values3
sns.distplot(Value)
```

Построим гистограмму распределения для суммы случайных величин: $\Sigma^{10}_{i=1} X_{i}$

```python
sum0 = 0
for i in range (1, 11):
    X = stats.uniform(0,1)
    Values = X.rvs(10000000)
    sum0 += Values
sns.distplot(sum0)
```

Построим гистограмму распределения для суммы случайных величин: $\Sigma^{50}_{i=1} X_{i}$

```python
sum0 = 0
for i in range (1, 51):
    X = stats.uniform(0,1)
    Values = X.rvs(10000000)
    sum0 += Values
sns.distplot(sum0)
```

Можно заметить, что при увеличении N,распределение суммы случайных величин стремится к нормальному

### Пункт "2"

Оценим вероятность $P(X_{1}+X_{2}+X_{3}>2)$ 

Данная вероятность представляет собой отношение количества "благоприятных" событий ($X_{1}+X_{2}+X_{3}>2$) к общему количеству событий

#### Сложный способ

```python
X1 = stats.uniform(0,1)
values1 = X1.rvs(10000000)
X2 = stats.uniform(0,1)
values2 = X2.rvs(10000000)
X3 = stats.uniform(0,1)
values3 = X3.rvs(10000000)
Value = values1 + values2 + values3
c = 0
for i in range (len(Value)):
    if Value[i] > 2:
        c += 1
prob = c/(len(Value))
prob
```

#### Простой способ

```python
X1 = stats.uniform(0,1)
values1 = X1.rvs(10000000)
X2 = stats.uniform(0,1)
values2 = X2.rvs(10000000)
X3 = stats.uniform(0,1)
values3 = X3.rvs(10000000)
Value = values1 + values2 + values3
np.sum(Value>2)/(len(Value))
```

### Дано

Даны следующие вводные данные:

Начальный капитал равен 100.

Пусть вероятности следующие:

$p_{1a}$ = 49% – это вероятность выигрыша в игре А

$p_{2a}$ = 51% – это вероятность проигрыша в игре А

$p_{1b}$ = 74% – это вероятность выигрыша в игре B, если капитал не кратен 3

$p_{2b}$ = 26% – это вероятность проигрыша в игре B, если капитал не кратен 3

$p_{3b}$ = 9% – это вероятность выигрыша в игре B, если капитал кратен 3

$p_{4b}$ = 91% – это вероятность проигрыша в игре В, если капитал кратен 3

### Симуляция Игры "А"

```python
pa1 = 0.49
pa2 = 0.51
x = np.random.random(100000)
y = np.random.random(100000)
c = 100
lsta = []
ya = []
for i in range (100000):
    if  y[i] > pa1:
        c -=1
    elif y[i] <= pa1:
        c +=1
    lsta.append(c)
    ya.append(i)
print(c)
plt.plot(ya, lsta)
plt.ylabel('Earnings')
plt.xlabel('Number of Trials')
```

### Симуляция Игры "Б"

```python
pb_not3 = 0.74
pb_yes3 = 0.09

x = np.random.random(100000)
y = np.random.random(100000)
yb = []
lstb = []
c = 100
for i in range (100000):
    if c % 3 == 1 or c % 3 == 2:
        if y[i] > pb_not3:
            c -= 1
        elif y[i] <= pb_not3:
            c +=1
    elif c % 3 == 0:
        if y[i] > pb_yes3:
            c -= 1
        elif y[i] <= pb_yes3:
            c += 1
    lstb.append(c)
    yb.append(i)
print(c)
plt.plot(yb, lstb)
plt.ylabel('Earnings')
plt.xlabel('Number of Trials')
```

### Игра С (игрок рандомно выбирает игру А или Б)

```python
pa = 0.49
pb_n3 = 0.74
pb_3 = 0.09
x = np.random.random(100000)
y = np.random.random(100000)
lstr = []
yr = []
c = 100
for i in range (100000):
    if x[i] > 0.50:
        if  y[i] > pa:
            c -=1
        elif y[i] <= pa:
            c +=1
    elif x[i] < 0.50:
        if c % 3 == 1 or c % 3 == 2:
            if y[i] > pb_n3:
                c -= 1
            elif y[i] <= pb_n3:
                c +=1
        elif c % 3 == 0:
            if y[i] > pb_3:
                c -= 1
            elif y[i] <= pb_3:
                c += 1
    lstr.append(c)
    yr.append(i)   
print(c)
plt.plot(yr, lstr)
plt.xlabel('Number of Trials')
plt.ylabel('Earnings')
```

### Представление результатов

Представим результат на совместном графике

```python
plt.plot(ya, lsta, label = 'Game A')
plt.plot(yb, lstb, label = 'Game B')
plt.plot(yr, lstr, label = 'Random')

plt.xlabel('Number of Trials')
plt.ylabel('Earnings')
plt.legend(loc = 'lower left')
```

Из проведенных симуляций можно сделать следующие выводы:

1) При бесконечном количестве попыток, игра А - проигрышна, и с увеличением числа попыток, проигрыш только увеличивается

2) При бесконечном количестве попыток, игра Б - тоже проигрышна, и с увеличением числа попыток, проигрыш только увеличивается.

3) Однако, если он равновероятно выбирает между двумя играми,то при заданных вероятностях, он выигрывает. В долгосрочной перспективе, его выигрыш увеличивается
